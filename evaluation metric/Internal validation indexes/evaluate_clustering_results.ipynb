```python
import os
import numpy as np
from tqdm import tqdm
import pickle

from evaluate import evaluate
from utils.io import load_file, load_distance_matrix
```

```python
result_dir = '/scratch/yk1962/ABC2/result/clustering/cluster_result/labels'
name_index = load_file('name_index', '/scratch/yk1962/ABC2/result/')
# methods = ['silhouette_score', 'dunn_index', 'davies_bouldin_index']
methods = ['silhouette_score']
evaluation_dir = '/scratch/yk1962/ABC2/result/evaluation'
```

```python
distance_matrix_list = [(load_distance_matrix('chamfer_distance_sparse.npz', '/scratch/yk1962/ABC2/result/chamfer_distance_result'),
                         'chamfer distance'),
                        (load_distance_matrix('iou.npy', '/scratch/yk1962/ABC2/result/IoU/'),
                         'DoU distance')
                       ]
```

```python
def save_obj(obj, name, obj_dir, postfix=True):
    filepath = os.path.join(obj_dir, name)
    if postfix:
        filepath += '.pkl'
    with open(filepath, 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
evaluation_dict = load_file('evaluation_result', evaluation_dir, load_type='pkl')
evaluation_dict
# evaluation_dict = {}
```

```python
for result in tqdm(os.listdir(result_dir)):
    try:
        labels = np.load(os.path.join(result_dir, result))
    except:
        continue
    for method in methods:
        for distance_matrix, distance_type in distance_matrix_list:
            subject_name = 'cluster result name: {}, evaluation method: {}, distance type: {}'.format(result, method, distance_type)
            if subject_name in evaluation_dict:
                continue
            score = evaluate(method, distance_matrix, labels)
            evaluation_dict[subject_name] = score
#             save_obj(evaluation_dict, 'evaluation_result', evaluation_dir)
```
