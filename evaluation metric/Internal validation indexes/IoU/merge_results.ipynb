```python
import os
import numpy as np
from tqdm import tqdm
from iou_constants import iou_coo_args_dir, num_data, result_dir
```

```python
os.listdir(iou_coo_args_dir),
```

```python
sub_result_list = sorted(filter(lambda result: result.split('.')[-1] == 'npz', os.listdir(iou_coo_args_dir)),
                         key=lambda result: int(result.split("_")[-2]))
```

```python
current_start = 0
for iou_coo_args_name in sub_result_list:
    split_name = iou_coo_args_name.split("_")
    assert(int(split_name[-2]) == current_start)
    current_start = int(split_name[-1].split(".")[0])
    
assert(current_start == num_data)
```

```python
num_comp = num_data * (num_data-1) // 2
```

```python
start = 0
data = np.zeros(num_comp)
ii = np.zeros(num_comp, dtype=int)
jj = np.zeros(num_comp, dtype=int)

for iou_coo_args_name in tqdm(sub_result_list):
    coo_args = np.load(os.path.join(iou_coo_args_dir, iou_coo_args_name))
    sub_data = coo_args["data"]
    end = start + len(sub_data)
    assert((data[start:end] == 0.).all() == True)
    data[start:end] = sub_data
    ii[start:end] = coo_args["ii"]
    jj[start:end] = coo_args["jj"]
    start = end
```

```python
import scipy.sparse as sp
```

```python
iou = sp.coo_matrix((data, (ii, jj))).tocsr()
```

```python
result_path = os.path.join(result_dir, "iou.npy")
np.save(result_path, iou)
```

```python
matrix = iou.todense()
```
