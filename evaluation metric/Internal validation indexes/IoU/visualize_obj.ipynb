```python
data_list = sorted(os.listdir(data_dir))
```

```python
def get_obj(idx):
    return trimesh.load_mesh(os.path.join(data_dir, data_list[idx], "{}_normalized.off".format(data_list[idx]))).show()

def get_objs(idx_list):
    for idx in idx_list:
        get_obj(idx)
```
