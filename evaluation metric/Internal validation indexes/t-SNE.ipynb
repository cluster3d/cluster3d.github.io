```python
import numpy as np
import os
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from random import choices
```

```python
latent_space_dir = os.path.join(os.getcwd(), '../result/clustering/TSNE')
```

```python
result_dict = {'byol_latent_space.npy': 'Byol-k=32.npy', 'atlasnet_latent_space.npy': 'AtlasNet-k=32.npy'}
```

```python
os.listdir(latent_space_dir)
```

```python
for clustering_method in os.listdir(latent_space_dir):
    path = os.path.join(latent_space_dir, clustering_method)
    latent_space = np.load(path)
    embedded = TSNE().fit_transform(latent_space)
    cluster_result_dir = os.path.join(os.getcwd(), '../result/clustering/cluster_result/labels')
    path = os.path.join(cluster_result_dir, result_dict[clustering_method])
    labels = np.load(path)
    num_clusters = max(labels) + 1
    indexes = np.array(list(range(22968)))
    chosen_indexes = choices(indexes, k=2000)
    chosed_labels = labels[chosen_indexes]
    cmap = plt.cm.get_cmap('hsv', num_clusters)
    for cluster_idx in range(num_clusters):
        cluster = np.where(chosed_labels == cluster_idx)[0]
        if len(cluster) == 0:
            continue
        plt.scatter(embedded[cluster, 0], embedded[cluster, 1],s=1, facecolor=cmap(cluster_idx))
    plt.title(result_dict[clustering_method])
    plt.show()
```
